```python
import numpy as np
import random
import matplotlib.pyplot as plt
from mpmath import *
import math
import cmath
def E(epsilon_1,epsilon_2,gamma,z1,center,delta):
    A = 30
    B = 40
    z = z1 - center
    u = math.exp(z/delta)
    r = 0
    psi_u = u*(1j * math.sqrt(B)) * hyp2f1(-math.sqrt(A) - 1j * math.sqrt(B), math.sqrt(A) - 1j * math.sqrt(B), 1 - 2j * math.sqrt(B), u) + r * (u*(-1j * math.sqrt(B))) * hyp2f1(-math.sqrt(A) + 1j * math.sqrt(B), math.sqrt(A) + 1j * math.sqrt(B), 1 + 2j * math.sqrt(B), u)
    epsilon_z = complex(epsilon_1*epsilon_2*((1-math.exp(z/delta))/(epsilon_1 - (epsilon_2*math.exp(z/delta)))),gamma/(1-((epsilon_1/epsilon_2)*math.exp(z/delta))))
    return ((abs(psi_u)/abs(epsilon_z))**2)*abs(epsilon_z)

z_value = np.arange(start = -0.015,stop = 0.015,step = 0.00001)
input = z_value.tolist()
i = 0
epsilon_1 = 5.28
epsilon_2 = -4.8
gamma = 0.00024
delta = 0.02
output_ratio = []
 
while i<3000:
    output_ratio.append(E(epsilon_1,epsilon_2,gamma,input[i],0,delta))
    i = i+1
#plt.scatter(input,output_ratio,s=4, label = 'Al2O3/Ag Nanowires')
#plt.plot(input,output_ratio,color='black',label = 'Al2O3/Ag Nanowires')
plt.semilogy(input, output_ratio)
plt.xlabel( u"\u03bcm")
plt.ylabel("I(z)/I(-inf)")
plt.legend()
print(E(epsilon_1,epsilon_2,gamma,0,0,delta))
plt.show()
```
